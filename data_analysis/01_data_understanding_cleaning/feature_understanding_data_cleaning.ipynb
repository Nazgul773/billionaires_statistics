## Check for missing data

```python
# Check for zero values first (they could indicate missing data)
(df == 0).sum()
# selfMade has a boolean dt -> zero values mean "False" here, so it's not a missing data related problem
# Otherwise, if there would be numeric columns with zeros, this should get checked, and if it's related to missing data, we can replace them with NaN's to handle them in the following with other potential NaN-values
```

```python
def get_missing_data_rate(input_df: pd.DataFrame):
    total = input_df.isna().sum().sort_values(ascending=False)
    percent = (input_df.isna().sum() / input_df.isna().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    return missing_data
```

```python
get_missing_data_rate(df)
```

```python
# Select only relevant features for the specific task + drop columns when more than 15 % of the data is missing
filtered_df = df[['rank', 'finalWorth', 'personName',
                  #   'source', 'countryOfCitizenship', 'organization','lastName', 'firstName',
                  'age', 'country', 'city', 'industries',
                  'selfMade', 'status', 'gender',
                  #'birthDate', 'title', 'state', 'residenceStateRegion', 'birthYear', 'date', 'category'
                  #'birthMonth', 'birthDay', 'cpi_country', 'cpi_change_country',
                  'life_expectancy_country',
                  #'tax_revenue_country_country',
                  #'latitude_country', 'longitude_country'
                  'gross_tertiary_education_enrollment', 'gross_primary_education_enrollment_country',
                  'gdp_country', 'total_tax_rate_country', 'population_country'
                  ]]
```

```python
get_missing_data_rate(filtered_df)
```

```python
# There are a few options to handle remaining missing data
# We can just drop the remaining rows with missing data 
# Otherwise we could fill the NaN-entries for categorical columns with the modus, and for numerical columns with mean/median of the specific column to avoid data-loss (if the dataset is small)
# We can work with mean if the data is not skewed (no outliers)
# Otherwise we'd choose median as the NaN-fill-option because it's robust against outliers
# To avoid data skewness we just drop the remaining rows with missing data here
filtered_df = filtered_df.dropna()
```

```python
# Check the last time if there's no missing data left
filtered_df.isna().sum().max()
```

## Check for outliers

```python
# Usually you can check for outliers through a boxplot related to the specific numerical (x) and a dependent categorical (y) feature
# But this type of dataset is very susceptible to outliers because of the wide range of billionaires' wealth
plt.figure(figsize=(14,6))
boxplot = sns.boxplot(data=filtered_df, x='finalWorth',y='industries')
boxplot.set(xlabel= 'Final Worth (M)', ylabel   = 'Industry')
plt.show()
```

```python
# To demonstrate, how handling outliers could be done, we drop all the rows where the billionaire's final worth is more than 100.000 Millions
# Define the condition to filter rows
condition = (filtered_df['finalWorth'] > 100000)
# Use boolean indexing to select rows that don't match the condition
filtered_df_no_outliers = filtered_df.loc[~condition]
filtered_df_no_outliers 
```

```python
# Check if we have any outliers left based on dropping-condition
filtered_df_no_outliers.loc[condition]
```
